Setting up the environment:

````python
import numpy as np
import matplotlib.pyplot as plt

from math import *
from utils import *

plt.rcParams['figure.dpi'] = 120 
````

Reading the data

````python
linear_data = readData("linear_function_data.txt")
X = linear_data[0, :]
F = linear_data[1, :]
````

Deriving the linear function with ```numpy```'s method for least squares

````python
A = np.vstack([X, np.ones(len(X))]).T
m, c = np.linalg.lstsq(A, F, rcond=10)[0]
````

````python
fig = plt.figure()
fig, ax = plt.subplots()

ax.set_title("linear fitted curve")

ax.plot(X, F, 'o', label='Original Data', markersize=5)
ax.plot(X, m*X, 'o', label='Fitted Data', markersize= 2)

ax.legend()

ax.set_xlabel("x")
ax.set_ylabel("f(x)")

fig.savefig('images/task1_1_linear.png')
````

Calculating the error

````python
squared_error = calc_error(m*X, F)**2
squared_error
````
