```python
# using the program from Exercise 3 to plot the Lorenz Attractor
from lorenz import *
from utils import *
plt.rcParams['figure.dpi'] = 120 

#parameters
sigma = 10
rho = 28
beta = 8./3.

dt = 0.01
end_time = 300
t = np.linspace(1, end_time, num=1000)

# plot the Lorenz attractor with staring values (10.0, 10.0, 10.0) and extract the x, y, z axis
x, y, z = plot_lorenz(sigma, rho, beta, dt, end_time)
```

```python
# shift the input data by delta_t lines to the left
delta_t = [10, 20, 30, 40]

# setup shifted data
x_shifted_0 = shift_data(x, delta_t[0])
x_shifted_1 = shift_data(x, delta_t[1])
x_shifted_2 = shift_data(x, delta_t[2])
x_shifted_3 = shift_data(x, delta_t[3])

shifted_data = [x_shifted_0, x_shifted_1, x_shifted_2, x_shifted_3]

# plot the first coordinate againt delayed versions (3 delayed axis total) 
# for different combinations of axis 

fig, axs = plt.subplots(2,2,figsize=(12,12), subplot_kw=dict(projection='3d'))
for i,ax in enumerate(axs.flat):
    idx1, idx2, idx3 = (i-1)%4, i, (i+1)%4
    ax.scatter3D(shifted_data[idx1],shifted_data[idx2], shifted_data[idx3], c=shifted_data[idx2], cmap='viridis')
    ax.set_xlabel('idx1')
    ax.set_ylabel('idx2')
    ax.set_zlabel('idx3')
    ax.set_title("shifted time step increments: %s, %s, %s" % (idx1, idx2, idx3))  
```

```python
# shift the input data by delta_t lines to the left
delta_t = [20, 40, 60, 80]

# setup shifted data
z_shifted_0 = shift_data(z, delta_t[0])
z_shifted_1 = shift_data(z, delta_t[1])
z_shifted_2 = shift_data(z, delta_t[2])
z_shifted_3 = shift_data(z, delta_t[3])

shifted_data = [z_shifted_0, z_shifted_1, z_shifted_2, z_shifted_3]

# plot the first coordinate againt delayed versions (3 delayed axis total) 
# for different combinations of axis 

fig, axs = plt.subplots(2,2,figsize=(12,12), subplot_kw=dict(projection='3d'))
for i,ax in enumerate(axs.flat):
    idx1, idx2, idx3 = (i-1)%4, i, (i+1)%4
    ax.scatter3D(shifted_data[idx1],shifted_data[idx2], shifted_data[idx3], c=shifted_data[idx2], cmap='viridis')
    ax.set_xlabel('idx1')
    ax.set_ylabel('idx2')
    ax.set_zlabel('idx3')
    ax.set_title("shifted time step increments: %s, %s, %s" % (idx1, idx2, idx3)) 
```
